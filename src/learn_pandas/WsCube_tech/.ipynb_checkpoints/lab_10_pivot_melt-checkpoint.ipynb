# Pivot and Melt Funtion

## Melt

```python
import pandas as pd
```

```python
df_1 = pd.DataFrame({
    "days": [1, 2, 3, 4, 5, 6],
    "eng": [10, 12, 14, 15, 16, 12], 
    "maths": [17, 18, 19, 13, 14, 16]
})

df_1
```

```python
pd.melt(df_1) # change all column from data frame to structure: variable - value
```

```python
pd.melt(df_1, id_vars=["eng"]) # keep column eng as id
```

```python
pd.melt(df_1, id_vars=["days"]) # keep column days as id
```

```python
pd.melt(df_1, id_vars=["days"], var_name="python") # keep column days as id, set variable column name
```

```python
pd.melt(df_1, id_vars=["days"], var_name="python", value_name="col_value") # keep column days as id, set variable column name, set value column name
```

## Pivot

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "days": [1, 2, 3, 4, 5, 6],
    "st_name": ["a", "b", "c", "a", "b", "c"],
    "eng":[10, 12, 14, 15, 16, 12],
    "maths": [17, 18, 19, 13, 14, 16]
})

df
```

```python
df.pivot(index="days", columns="st_name")
```

```python
df_1 = pd.DataFrame({
    "days": [1, 2, 3, 4, 5, 6],
    "st_name": ["a", "b", "a", "a", "b", "b"],
    "eng":[10, 12, 14, 15, 16, 12],
    "maths": [17, 18, 19, 13, 14, 16]
})
# df_1
df_1.pivot(index="days", columns="st_name", values=["eng"])
```

```python
df_1 = pd.DataFrame({
    "days": [1,1,1,1,2,2],
    "st_name": ["a", "b", "a", "a", "b", "b"],
    "eng":[10, 12, 14, 15, 16, 12],
    "maths": [17, 18, 19, 13, 14, 16]
})
df_1.pivot_table(index="st_name", columns="days", aggfunc="mean")
```

```python
df_1 = pd.DataFrame({
    "days": [1,1,1,1,2,2],
    "st_name": ["a", "b", "a", "a", "b", "b"],
    "eng":[10, 12, 14, 15, 16, 12],
    "maths": [17, 18, 19, 13, 14, 16]
})
df_1.pivot_table(index="st_name", columns="days", aggfunc="sum")
```

```python
df_1 = pd.DataFrame({
    "days": [1,1,1,1,2,2],
    "st_name": ["a", "b", "a", "a", "b", "b"],
    "eng":[10, 12, 14, 15, 16, 12],
    "maths": [17, 18, 19, 13, 14, 16]
})
df_1.pivot_table(index="st_name", columns="days", aggfunc="sum", margins=True)
```

```python
df_1 = pd.DataFrame({
    "days": [1,1,1,1,2,2],
    "st_name": ["a", "b", "a", "a", "b", "b"],
    "eng":[10, 12, 14, 15, 16, 12],
    "maths": [17, 18, 19, 13, 14, 16]
})
df_1.pivot_table(index="st_name", columns="days", aggfunc="mean", margins=True)
```
