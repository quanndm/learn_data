# Handle Missing Data
- dropna
- fillna

## dropna

```python
import pandas as pd
```

```python
df = pd.read_csv("./file_import/test_missing_data.csv")
df
```

```python
df.dropna() # delete row if values in row is NaN
```

```python
df.dropna(axis=1) # same above but delete column because axis=1
```

```python
df_1 = pd.read_csv("./file_import/test_missing_data_1.csv")
df_1 
```

```python
df_1.dropna(how="any") # delete any row if values on that row is NaN
```

```python
df_1.dropna(how="all") # only remove row if all value on that row is NaN
```

```python
df.dropna(subset=["CHANGE"]) # only check value on column CHANGE, if values on that column is NaN, it will drop that column
```

```python
df.dropna(
    #inplace=True 
)
# will drop and replace data frame
```

## fillna

```python
df
```

```python
df.fillna("python") # it will fill python for any field has NaN value
```

```python
df.ffill() # fill NaN value with forward valid value
```

```python
df.bfill() # fill NaN value with previous valid value
```

```python
df.bfill(axis=1)
```
