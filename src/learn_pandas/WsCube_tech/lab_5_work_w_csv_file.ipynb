# Working with CSV

## Write CSV

```python
import pandas as pd
```

```python
dic = {
    "a" : [1,2,3,4,5,6],
    "s" : [1,2,3,4,5,6],
    "d" : [1,2,3,4,5,6]
}
df_1 = pd.DataFrame(dic)
df_1
```

```python
import os

folder_name = "file_exported"

isExist = os.path.exists(folder_name)
if not isExist:
    os.makedirs(folder_name)

df_1.to_csv(f"./{folder_name}/Test_new.csv")
```

```python

folder_name = "file_exported"

isExist = os.path.exists(folder_name)
if not isExist:
    os.makedirs(folder_name)

df_1.to_csv(f"./{folder_name}/Test_new.csv", index=False)
```

## Read CSV

```python
import pandas as pd
import os
folder_import = "file_import"
```

```python
data_csv_2 = pd.read_csv(f'./{folder_import}/test.csv', nrows=1) # get 1 row from the first
print(data_csv_2)
print(type(data_csv_2))
```

```python
data_csv_3 = pd.read_csv(f'./{folder_import}/test.csv', usecols=["SYMBOL", "CHANGE"]) # select column symbol and change
data_csv_3
```

```python
data_csv_4 = pd.read_csv(f'./{folder_import}/test.csv', skiprows=[0]) 
# get date from row 2 to n, values of row 1 become header of data set, it receive a list of row want to skip
data_csv_4
```

```python
data_csv_5 = pd.read_csv(f'./{folder_import}/test.csv', index_col="SYMBOL") # declare a column as a index
data_csv_5
```
